```python
class Brand:
    def __init__(self,bno,bname,brate,bcat,bprice):
        self.bno=bno
        self.bname=bname
        self.brate=brate
        self.bcat=bcat
        self.bprice=bprice
class orgnisationManagement:
    def __init__(self,ogname,brandlist):
        self.ogname=ogname
        self.brandlist=brandlist
        
    def CategoriseBrandsAlp(self):
        a={}
        for i in self.brandlist:
            if i.bprice>=1 and i.bprice<=39:
                a.setdefault("1-39",[]).append(i.bname)
            elif i.bprice>=40 and i.bprice<=59:
                a.setdefault("40-59",[]).append(i.bname)
            elif i.bprice>=60 and i.bprice<=100:
                a.setdefault("60-100",[]).append(i.bname)
        alist=sorted(a.items(),key=lambda x: x[1])
        sortdict={k:v for k,v in alist}
        for j in sortdict.values():
            j=tuple(j)
        return sortdict
                
        
    def SecondHighestRatedBrand(self,brandcat):
        l1=[]
        for i in self.brandlist:
            if i.bcat==brandcat:
                l1.append(i.brate)
                
        if len(l1)<2:
            return None
        l2=list(set(l1))
        l2.sort(reverse=True)
        for i in self.brandlist:
            if i.bcat==brandcat and i.brate==l2[1]:
                return i.bname
                
            
        
        
        
        
```
